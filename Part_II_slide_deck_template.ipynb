## Dataset Overview

> This data set contains 9,994 rows with each row being an information about each passengers and 21 variables which describes the product/customer. The dataset contains features like Segment, Category, Product Name, State, Profit, and many others.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Superstore.csv', engine='python')
````

> Note that the above cells have been set as "Skip"-type slides. That means
that when the notebook is rendered as http slides, those cells won't show up.

## Regions with the highest Profit

> This visualization will help us to know which regions performed the best in terms of Profit generation. The visaulization made us to understand that the west generated more profit, followed by the east.



````python
sb.barplot(x = 'Region',y = 'Profit',data = df)
plt.xticks(rotation = 90)
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profits by each Region');
````

## Segment and Category with the Highest Profit

> This visualization will help us know the segments and category that generates the highest profits. The Technology Category generated most profits across all segments

````python
plt.figure(figsize = [10, 6])
plt.title('Profits by Segment and Categories')
plt.xticks(rotation = 90)
sb.barplot(data = df, x = 'Segment', y = 'Profit', hue = "Category");
````

## States with the Highest Profit

> This visualization will help us know the states where the highest profits are genrated. Going by the visualization, i discovered that Indiana and Washington generated more profits than New york and California where the company has lots of customers

````python
plt.figure(figsize = [10, 6])
plt.title('Top 5 States with Highest Profits')
plt.xticks(rotation = 90)
sb.barplot(data = df.sort_values('Profit',ascending = False).head(5), x = 'State', y = 'Profit');
````

## Cities with the Highest Profit

> This visualization will help us know the top 5 cities where the highest profits are generated. As we can see, Lafayette, Seattle and Newark are the leading cities in terms of profits

````python
plt.figure(figsize = [10, 6])
plt.title('Top 5 Cities with Highest Profits')
plt.xticks(rotation = 90)
sb.barplot(data = df.sort_values('Profit',ascending = False).head(5), x = 'City', y = 'Profit');
````
